A wordcloud shows a visual of the frequencies of words in a text.  Words are arranged in some artistic manner (usually), with the size of the font scaled to represent how common the word is.  Larger font size means a word occurs more frequently in the document or corpus. The color is not signficant in most word clouds, only font size. Often bad word clouds don't remove stop words or other frequent terms that would be "obvious."  They are also criticized by many for hiding context (obviously, because they are just separate words).

## Word Clouds show relative frequency of vocab tokens in a text

```python
# increasing the size of the figure drawn (width, height)
from pylab import rcParams
rcParams['figure.figsize'] = (20,15)
```

```python
tokens = nlp.tokenize_text("data/books/Austen_Emma.txt")
```

#### The code to make the wordcloud requires a string, not a list of tokens.  It does its own tokenization. But we want to have more control, so we can clean it, then convert it to a string.

How do you make a string out of a list of strings?  Use join!

```python
text = ' '.join(tokens)
```

```python
def make_word_cloud(text, stops=None):
    import wordcloud
    # you have to figure out a font path for your machine (sigh)
    myWordcloud = wordcloud.WordCloud(font_path="data/Lato-Medium.ttf", 
                                      stopwords=stops).generate(text)
    plt.imshow(myWordcloud)
    plt.axis("off")
    plt.show()
```

```python
# giving it the text we joined above...
make_word_cloud(text)
```

Wordcloud also allows passing in a custom stopwords list (it just uses its own otherwise); may want to join yours with the english one and use those, or else clean your tokens before making the text, using you own functions.  Here it is not cleaned, and not using any stop word list except a custom list I passed in:

```python
make_word_cloud(text, stops=["Mr", "Mis"])
```

Hint: To clean your own tokens, you would run clean_tokens on your tokens before doing "join." You can also remove your own custom stops from the argument to make_word_cloud as I did above.
